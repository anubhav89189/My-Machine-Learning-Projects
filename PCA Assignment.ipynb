```python
#Pick first 15K data-points to work on for time-effeciency.
#Excercise: Perform the same analysis on all of 42K data-points.

labels= l[0:15000]           #labels with 15k data points
data = d[0:15000]               #data with 15k data points

print(labels.shape)
print(data.shape)
```

```python
# import standard scalar
from sklearn.preprocessing import StandardScaler

SC= StandardScaler()

#fit transform data
standardized_data = SC.fit_transform(data)


#print shape of standardized_data
print(standardized_data.shape)
```

```python
sample_data = standardized_data[0]
sample_data



```

```python
#use matrix multiplication using numpy to find covariance matrix
covar_matrix = np.dot(sample_data.T, sample_data)
covar_matrix
```

```python
#print shape of covar_matrix
print(covar_matrix.shape)
```

```python
# finding the top two eigen-values and corresponding eigen-vectors
# for projecting onto a 2-Dim space.

# Correctly compute the covariance matrix from the standardized data.
# 'rowvar=False' specifies that columns are variables and rows are observations.
covar_matrix = np.cov(standardized_data, rowvar=False)

# The 'eigvals' argument in np.linalg.eigh is causing a TypeError,
# which suggests an older NumPy version. We will compute all
# eigenvalues and eigenvectors and then select the top two.
# np.linalg.eigh returns eigenvalues in ascending order.
values_all, vectors_all = np.linalg.eigh(covar_matrix)

# Select the last two eigenvalues (largest) and their corresponding eigenvectors.
# The original code's intent was to have the eigenvector for the 2nd largest
# eigenvalue at index 0, and the eigenvector for the largest eigenvalue at index 1.
# So we select columns -2 (2nd largest) and -1 (largest) from vectors_all.
values = values_all[-2:]
vectors = vectors_all[:, [-2, -1]] # This ensures vectors has shape (784, 2) and desired order.

print(vectors.shape)

# here the vectors[:, 1] represent the eigen vector corresponding 1st principal eigen vector
# here the vectors[:, 0] represent the eigen vector corresponding 2nd principal eigen vector
```

```python
# projecting the original data sample on the plane
#formed by multiplication of two principal eigen vectors with transposed sample_data




# multiplication of two principal eigen vectors with transposed sample_data to get 2d projected data
new_coordinates = np.dot(standardized_data, vectors)

# print resultant new data points
print(new_coordinates.shape)
print(new_coordinates)
```

```python
import pandas as pd

# appending label to the 2d projected data
new_coordinates = pd.DataFrame(new_coordinates)
new_coordinates['label'] = labels
new_coordinates


# creating a new data frame for ploting the labeled points.
dataframe = pd.DataFrame(data=new_coordinates)
dataframe

# print head
dataframe.head()
```

```python
# ploting the 2d data points with seaborn
import seaborn as sn
```

```python
# import decomposition
import sklearn.decomposition as skldec

# initializing the pca
pca = skldec.PCA()
```

```python
# the number of components = 2
pca.n_components = 2

# fit transform sample data using pca
pca_data = pca.fit_transform(standardized_data)
pca_data

# pca_reduced will contain the 2-d projects of simple data
pca_data
```

```python
# the number of components = 2
pca.n_components = 2

# Re-fit and transform standardized_data to ensure pca_data has only 2 components
pca_data = pca.fit_transform(standardized_data)

# attaching the label for each 2-d data point
pca_data = np.concatenate((pca_data, labels.values.reshape(-1, 1)), axis=1)


# creating a new data fram which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "label"))
pca_df

# plot using facetgrid
sn.FacetGrid(pca_df, hue='label', height=6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend()
plt.show()
```

```python

# the number of components = 784
pca.n_components = 784


# fit transform sample data using pca
pca_data = pca.fit_transform(standardized_data)
pca_data



#calculating percentage of variance explained in the data
percentage_var_explained = pca.explained_variance_ratio_
percentage_var_explained



#cumulative sum of the percentage_var_explained
cum_var_explained = np.cumsum(percentage_var_explained)
cum_var_explained


# Plot the PCA spectrum
plt.figure(figsize=(8, 6))
plt.bar(range(784), percentage_var_explained, alpha=0.5, align='center', label='Individual explained variance')
plt.step(range(784), cum_var_explained, where='mid', label='Cumulative explained variance')
plt.grid()
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
```
